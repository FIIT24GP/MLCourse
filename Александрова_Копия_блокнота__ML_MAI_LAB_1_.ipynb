### Что такое IQR и как его использовать для обработки выбросов:

**IQR (Interquartile Range)**, или межквартильный размах, это мера статистического разброса и используется для определения выбросов. IQR рассчитывается как разность между 75-м и 25-м процентилями данных (Q3 и Q1 соответственно).

Выбросы часто определяются как значения, которые находятся за пределами следующих границ:
- Нижняя граница: Q1 - 1.5 * IQR
- Верхняя граница: Q3 + 1.5 * IQR

#### Примерный код на Python с использованием pandas:

```python
import pandas as pd

# Загрузка данных
df = pd.read_csv('path_to_file.csv')

# Рассчитываем IQR для колонки 'example_column'
Q1 = df['example_column'].quantile(0.25)
Q3 = df['example_column'].quantile(0.75)
IQR = Q3 - Q1

# Определяем границы для выбросов
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Фильтрация данных для удаления или корректировки выбросов
filtered_df = df[(df['example_column'] >= lower_bound) & (df['example_column'] <= upper_bound)]
```

````python
# заведем для красоты печати - добиваем название до максимальной длины названия
def add_spaces(input_string, total_length):
    # Проверяем, если строка уже больше или равна нужной длине, ничего не делаем
    if len(input_string) >= total_length:
        return input_string
    # Иначе добавляем пробелы до нужной длины
    return input_string + ' ' * (total_length - len(input_string))

# format number with spaces
def fnum_sp(number, total_length, decimal_places):
    # Форматируем число с нужным количеством знаков после запятой
    formatted_number = f"{number:.{decimal_places}f}"
    # Проверяем, если строка уже больше или равна нужной длине, ничего не делаем
    if len(formatted_number) >= total_length:
        return formatted_number
    # Иначе добавляем пробелы до нужной длины
    return formatted_number + ' ' * (total_length - len(formatted_number))
````

````python
# заведем еще один датафрейм из изначального - проверим потом
# насколько сильно повлияло это удение
# Сохраняем копию исходных данных для дальнейшего сравнения
before_numeric_data = numeric_data.copy()

# Создаем пустой DataFrame для хранения статистик до очистки
before_allstatdata_df = pd.DataFrame()

# Функция для расчета статистик
def MainStatData(col_name, data):
    mean_value = data[col_name].mean()
    median_value = data[col_name].median()
    min_value = data[col_name].min()
    max_value = data[col_name].max()
    mean_abs_dev = data[col_name].apply(lambda x: abs(x - mean_value)).mean()
    variance = data[col_name].var()

    return {"Column": col_name, "Mean": mean_value, "Median": median_value, "Min": min_value,
            "Max": max_value, "MeanAbsDev": mean_abs_dev, "Variance": variance}

# Рассчитываем статистики для данных до очистки (before_numeric_data)
for col in before_numeric_data.columns:
    before_allstatdata_df = pd.concat([before_allstatdata_df, pd.DataFrame([MainStatData(col, before_numeric_data)])], ignore_index=True)
````

````python
# заведем функцию для расчета этого показателя IQR
# пока чисто посмотрим что и как
def IQR(col_name):
  Q1 = numeric_data[col_name].quantile(0.25)
  Q3 = numeric_data[col_name].quantile(0.75)
  IQR = Q3 - Q1

  # Определяем границы для выбросов
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
# выведем - должны же мы гдето посмотреть
  print (f'Колонка {add_spaces(col_name,7)}: Q1 = {fnum_sp(Q1,7,3)}, \
  Q3 = {fnum_sp(Q3,7,3)}, IQR = {fnum_sp(IQR,7,3)}, \
  lower_bound = {fnum_sp(lower_bound,8,3)},\
  upper_bound = {fnum_sp(lower_bound,7,3)}')

for my_colname in numeric_data.columns.tolist():
  IQR(my_colname)
````

Вот теперь все понятно, можно делать основной код

Модифицируем функцию которая будет считать показатель и сведем расчеты в датафрейм, чтобы удобно было сортировать и выуживать имена колонок

````python
# Создаем пустой DataFrame с нужными столбцами
all_iqr_data = pd.DataFrame(columns=["Column", "Q1", "Q3", "IQR", "LowerB", "UpperB", "OutProc"])

# Функция для расчета IQR и возврата данных в виде словаря
def NewIQR(col_name):
    Q1 = numeric_data[col_name].quantile(0.25)
    Q3 = numeric_data[col_name].quantile(0.75)
    IQR_value = Q3 - Q1

    # Определяем границы для выбросов
    lower_bound = Q1 - 1.5 * IQR_value
    upper_bound = Q3 + 1.5 * IQR_value

    out_percent = (numeric_data[(numeric_data[col_name] < lower_bound) | (numeric_data[col_name] > upper_bound)].shape[0] / numeric_data[col_name].shape[0]) * 100

    # Возвращаем словарь с именем колонки и расчетами
    return {
        "Column": col_name,
        "Q1": Q1,
        "Q3": Q3,
        "IQR": IQR_value,
        "LowerB": lower_bound,
        "UpperB": upper_bound,
        "OutProc": out_percent
    }

# не округляем а просто показываем
pd.set_option('display.float_format', '{:.3f}'.format)

# Проходим по всем колонкам и добавляем данные в DataFrame
for my_colname in numeric_data.columns.tolist():
    iqr_data = NewIQR(my_colname)
    # Используем loc для добавления строки в DataFrame
    all_iqr_data.loc[len(all_iqr_data)] = iqr_data

all_iqr_data = all_iqr_data.sort_values(by='IQR', ascending=False)

# Выводим
all_iqr_data
````

Поскольку получилось что на один только показатель IQR надеяться не стоит, то получается так - где нулевой процент там и корректровать получается нечего, поскольку все занчения лежат внутри интервала

Будем придерживаться изначального плана по IQR и количеству значений. Где количество занчений меньше 5% - удалим, где более там скорректируем заменив на среднее.

````python
# Проходим по всем результатам в all_iqr_data
for _, iqr_data in all_iqr_data.iterrows():
    # Инициализируем счетчики удаленных и замененных значений
    del_val = 0
    ch_val = 0
    col_name = iqr_data["Column"]
    lower_bound = iqr_data["LowerB"]
    upper_bound = iqr_data["UpperB"]
    out_percent = iqr_data["OutProc"]

    if out_percent < 5:
        # Удаляем строки с выбросами
        numeric_data = numeric_data[(numeric_data[col_name] >= lower_bound) & (numeric_data[col_name] <= upper_bound)]
        del_val += 1
    else:
        # Заменяем выбросы на среднее значение по колонке
        mean_value = numeric_data[col_name].mean()
        numeric_data[col_name] = np.where((numeric_data[col_name] < lower_bound) | (numeric_data[col_name] > upper_bound),
                                          mean_value, numeric_data[col_name])
        ch_val += 1

    print(f"Колонка {add_spaces(col_name,7)}, удалили {del_val}, заменили {ch_val}")
````

Кажется получилась какая то ерунда. Почти все так называемые выбросы оказались внутри интервала Q1 Q3. Удалили всего по одному значению. Но все сделано по инструкции.

Выполним дополнительные вычисления для письменного отчета - сильно ли отличается среднее от медианы, как сильно друг от друга удалены минимум и максимум

````python

# Функция для расчета показателей
def MainStatData(col_name):
    mean_value = numeric_data[col_name].mean()
    median_value = numeric_data[col_name].median()
    min_value = numeric_data[col_name].min()
    max_value = numeric_data[col_name].max()
    mean_abs_dev = numeric_data[col_name].apply(lambda x: abs(x - mean_value)).mean()  # Среднее отклонение
    variance = numeric_data[col_name].var()  # Дисперсия

    # Возвращаем строку данных
    return {"Column": col_name, "Mean": mean_value, "Median": median_value, "Min": min_value,
            "Max": max_value, "MeanAbsDev": mean_abs_dev, "Variance": variance}

# Создаем пустой DataFrame для хранения всех результатов
allstatdata_df = pd.DataFrame()

# Проходим по всем колонкам и сохраняем данные в DataFrame
for my_colname in numeric_data.columns.tolist():
    st_data = MainStatData(my_colname)
    allstatdata_df = pd.concat([allstatdata_df, pd.DataFrame([st_data])], ignore_index=True)

# Разница между средним и медианой
allstatdata_df['Mean_Median_Diff'] = (allstatdata_df['Mean'] - allstatdata_df['Median']).abs()

# Разница между минимумом и максимумом
allstatdata_df['Min_Max_Diff'] = allstatdata_df['Max'] - allstatdata_df['Min']

allstatdata_df
````

````python
from scipy.stats import pearsonr

# Построение диаграммы рассеяния между RM и MEDV
def plot_scatter_rm_medv(df):
    plt.figure(figsize=(8, 6))
    plt.scatter(numeric_data['RM'], numeric_data['MEDV'], alpha=0.5, color='blue')
    plt.title('Диаграмма рассеяния: Количество комнат (RM) vs Стоимость жилья (MEDV)')
    plt.xlabel('Количество комнат (RM)')
    plt.ylabel('Стоимость жилья (MEDV)')
    plt.grid(True)
    plt.show()

# Расчет коэффициента корреляции Пирсона
correlation_coefficient, p_value = pearsonr(numeric_data['RM'], numeric_data['MEDV'])

# Вывод коэффициента корреляции
print(f"Коэффициент корреляции Пирсона: {correlation_coefficient:.3f}")
print(f"P-значение: {p_value:.3f}")

plot_scatter_rm_medv(numeric_data)
````

*Разница между корреляцией и причинно-следственной связью заключается в следующем:*

Корреляция — это статистическая взаимосвязь между двумя или более переменными. Коэффициенты корреляции не подразумевают наличие причинной зависимости, это лишь показатель того, что изменения одной переменной сопровождаются изменениями другой.

Причинно-следственная связь — это прямая взаимосвязь между двумя событиями, при которой одно событие (причина) приводит к возникновению другого события (следствия). Установления причинной связи определяются, как правило, экспериментальным путем.

Также обнаружение корреляции между двумя переменными не указывает на причинно-следственную зависимость. Чтобы подтвердить наличие причинной связи, необходимо проводить дополнительные исследования и контролируемые эксперименты.
